```python
from keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt

#Load data:
(X_train , y_train) , (X_test,y_test) = mnist.load_data()
```

```python
#Show
plt.Figure(figsize=(12,6))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_train[i] , cmap="gray")
    plt.axis("off")
    plt.title(f"Lable: {y_train[i]}")
plt.show()
```

```python
#Distribution of classes:
unique_nums , counts = np.unique(y_train , return_counts=True)
bars = plt.bar(unique_nums , counts )
plt.bar_label(bars, fmt='%d', padding=3)
plt.xlabel("Digit(1-9)")
plt.ylabel("Counts")
plt.title("Distribution of digits in training set")
plt.show()
```

```python
#Mean image for each digit:
plt.Figure(figsize=(12,6))

for i in range(10):
    tracked_digit_arr = X_train[y_train == i]                    #Filter to track down same arrays as same Feature
    tracked_digit_arr_mean = np.mean(tracked_digit_arr , axis=0)
    plt.subplot(2,5,i+1)
    plt.imshow(tracked_digit_arr_mean , cmap="gray")
    plt.axis("off")
plt.show()     
```
